```python
#Written in JupyterLab Version 3.6.3
#Python 3.10
#pip 23.1.2
import pandas as pd #Version 2.0.1
import altair as alt #Version 5.0.0
```

```python
#Load full_df file 
full_filepath = './data/cleaned_data/full_df.csv'
full_df = pd.read_csv(full_filepath)
full_df[['Region','Year','Home Size','Median Income - 2022 Dollars']].head()
```

```python
full_df['Owner-Occupied-Home-Proportion-dec'] = full_df['Owner-Occupied-Home-Proportion']/100

full_df['Year-Month'] = full_df['Year'].astype(str) + '-' + full_df['Month']
full_df['Year-Month'] = pd.to_datetime(full_df['Year-Month'], format='%Y-%B').dt.date
full_df['Year-Month'] = full_df['Year-Month'].astype(str)
chart = alt.Chart(full_df).mark_bar().encode(
    x=alt.X('Year-Month:T', axis=alt.Axis(labelAngle=90, labelFontWeight='bold'), title=''),
    y=alt.Y('average(Median Income - 2022 Dollars):Q', axis=alt.Axis(labelFontWeight='bold'), title=''),
    color='Owner-Occupied-Home-Proportion-dec:Q'
).properties(
    width=200,
    height=100
).facet( 
    column=alt.Column('Region:O', sort=['United States','Northeast','South','Midwest','West'], header=alt.Header(labelFontWeight='bold'), title=''), 
    title=alt.TitleParams('Regional Income and Home Ownership Trends', anchor='middle', baseline='line-top', lineHeight=1)
)
display(chart)
```

The graph above shows median regional income per year with a shading based on the proportion of people who own homes. Within each region, income steadily increased each year before peaking in 2019. Meanwhile, home ownership peaked in 2020. Generally, the trend seems to indicate an increase in home ownership and income over time.
