```python
import pandas as pd
import numpy as np
import calendar
```

```python
pd.set_option('display.max_rows', 5)

# Load the MHS Excel file
MHS_filepath = '/Users/joshhaber/Downloads/mhstabavgsls.xlsx' #Local filepath to replace with reproducible version 
MHS_df = pd.read_excel(MHS_filepath, header=[3,4],skipfooter=9) #Format header and remove footer
# Set the index to the first column, which is multilevel
MHS_df.set_index(MHS_df.columns[0], inplace=True)
mult_idx = MHS_df.index


# Display the DataFrame
MHS_df
```

```python
idx_list = mult_idx.to_list() #create index list and accumulators
full_year = None
years = []
months = []

for item in idx_list: #iterate through list
    if isinstance(item, int): #if item in list is int
        full_year = item #it's a year
    else:
        years.append(full_year) #if its not a year, store the year
        months.append(item) #store month
        
MHS_df = MHS_df.dropna() #the 10 year values with no month are all null        
        
MHS_df['Year'] = years
MHS_df['Month'] = months

MHS_df
```

```python
#Fully prepared MHS_df

# Stack the DataFrame to pivot the multi-level columns into rows
MHS_df_stacked = MHS_df.set_index(['Year', 'Month']).stack(level=[0, 1])

# Reset the index to convert the index labels into columns
MHS_df_stacked_reset = MHS_df_stacked.reset_index()

# Rename the columns
MHS_df_stacked_reset.columns = ['Year', 'Month', 'Region', 'Home Size', 'Average Sales Price']

# Display the transformed DataFrame
MHS_df_stacked_reset


```

```python
# Define the filepath
HIT_filepath = '/Users/joshhaber/Downloads/h06ar.xlsx'

# Load the Excel file, skipping the repeated header rows
HIT_df = pd.read_excel(HIT_filepath,skiprows=[0,1,2,3,4,5])

# Rename the 'United States' column to 'Region and year'
HIT_df.rename(columns={'United States': 'Region and year'}, inplace=True)

# Set the first row of the 'Region and year' column to 'United States'
HIT_df.loc[0, 'Region and year'] = 'United States'

HIT_df = HIT_df.dropna(subset=['Region and year', 'Unnamed: 1'])



# Filter out rows containing "2017 (40)"
HIT_df = HIT_df[HIT_df['Region and year'] != '2017 (40)']

# Filter out rows containing "2013 (38)"
HIT_df = HIT_df[HIT_df['Region and year'] != '2013 (38)']

HIT_df=HIT_df.dropna().reset_index(drop=True)

HIT_df.head()
```

```python
#Pair Regions with corresponding years

regions = ['United States', 'Northeast', 'Midwest', 'South', 'West']
HIT_years = range(1975, 2023)  # Adjusted range from 1975 to 2022
years_list = [str(year) for year in HIT_years]
time_len = len(years_list)
sorted_years_list = sorted(years_list, reverse=True)

# Map regions to HIT_df['Region and year']
for i, region in enumerate(regions):
    start_index = i * time_len
    end_index = start_index + time_len
    HIT_df.loc[start_index:end_index-1, 'Region and year'] = [f"{region}, {year}" for year in sorted_years_list]

HIT_df.head()
```

```python
#Name Columns and split the Region and Year column into two separate columns 

column_list = ['Region and Year','Number (Thousands)',
               'Median Income - Current Dollars','Median Income - 2022 Dollars',
               'Mean Income - Current Dollars','Mean Income - 2022 Dollars']

# Rename columns of HIT_df
HIT_df.columns = column_list

# Split 'Region and Year' into 'Region' and 'Year' columns
HIT_df[['Region', 'Year']] = HIT_df['Region and Year'].str.split(', ', expand=True)

# Drop the original 'Region and Year' column
HIT_df.drop(columns=['Region and Year'], inplace=True)



HIT_df.head()
```

```python
# Reorder columns with 'Region' and 'Year' on the far left
HIT_df.insert(0, 'Region', HIT_df.pop('Region'))
HIT_df.insert(1, 'Year', HIT_df.pop('Year'))
```

```python
# Assuming you have your DataFrame stored in 'HIT_df'
# Convert 'Year' column to numeric if it's not already
HIT_df['Year'] = pd.to_numeric(HIT_df['Year'])

# Create a new DataFrame to store the expanded data
expanded_df = pd.DataFrame()

# Iterate through each row in the original DataFrame
for index, row in HIT_df.iterrows():
    # Extract relevant information from the row
    region = row['Region']
    year = row['Year']
    # Iterate through each month and create a new row for each
    for month in range(1, 13):
        # Create a new row with the same data as the original row
        new_row = row.copy()
        # Assign the month name to the 'Month' column
        new_row['Month'] = calendar.month_name[month]
        # Append the new row to the expanded DataFrame
        expanded_df = expanded_df.append(new_row, ignore_index=True)

# Display the expanded DataFrame
expanded_df.head()
```

```python
# Filter the DataFrame to keep only rows with years between 2014 and 2023 to match the MHS_df
filtered_HIT_df = expanded_df[(expanded_df['Year'] >= 2014) & (expanded_df['Year'] <= 2023)]
```

```python
# Merge the two DataFrames on the shared columns 'Year' and 'Region' and 'Month'
secondary_df = pd.merge(MHS_df_stacked_reset, filtered_HIT_df, on=['Year', 'Region', 'Month'], how='inner')
secondary_df.rename(columns={'Number (Thousands)': 'Number of Households (Thousands)'}, inplace=True)
secondary_df
```

```python
# Save the DataFrame to a CSV file
secondary_df.to_csv('secondary_df.csv', index=False)
```
